```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('bmh')
import seaborn as sns
import os
```

```python
df= pd.read_csv(r'./HRDataset_v14.csv', sep=',')
print(df.shape)
df.head(2)
```

```python
mean_df = df['Salary'].mean()
print("**punto 1**")
print("**promedio salarios**")
print(mean_df)
print(df.describe())
```

```python
print("**punto 2**")
ax = sns.displot(data=df, x='Salary', fill=True)
ax.set(xlabel='Salario', ylabel='Cantidad', title='DistribuciÃ³n  de Salario')
plt.figure();
```

```python
print("**punto 3 Distribucion de salarios Departamento y Genero**")
ax = sns.catplot(data=df, kind='violin', x='Department', y='Salary', hue='Sex', split=True)
ax.set(xlabel='Departamento', ylabel='Rango Salarial', title='Distribucion de salarios Departamento y Genero');
```

```python
print("**punto 4 , Variantes**")
ax = sns.displot(data=df, x='DateofHire', fill=True)
ax.set(xlabel='Fecha de Contratacion', ylabel='Cantidad', title='Cantidad de contratados por fecha')
plt.figure();
```

```python
fecha = df['DateofHire']
fecha = pd.DatetimeIndex(fecha)
fecha
df['date'] = fecha
df['year'] = pd.DatetimeIndex(df['DateofHire']).year


df1 = df[['EmpID', 'year']]
df1.head()
df1=df1.groupby(['year'])
df1=df1.count().sort_values(['year'], ascending=False)
```

```python
ax = sns.displot(data=df, x='Sex', fill=True)
ax.set(xlabel='Genero', ylabel='Cantidad', title='Cantidad de contratados por Genero')
plt.figure();
```
